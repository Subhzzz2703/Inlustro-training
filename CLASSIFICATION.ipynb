```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
report=pd.read_csv("/content/column_2C_weka.csv")
```

```python
x=report.drop('class',axis=1)
y=report['class']
```

```python
lb=LabelEncoder()
y_encoded=lb.fit_transform(y)
print(y_encoded)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_encoded,test_size=0.3,random_state=42)
```

```python
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f'accuracy:{accuracy*100:.2f}%')
```

```python
decode=lb.inverse_transform(y_pred)
print(f'PREDICTED LABELS:{decode}')
```

```python
new_data=pd.DataFrame({
    'pelvic_incidence':[70.0],
    'pelvic_tilt numeric':[30.5],
    'lumbar_lordosis_angle':[50.6],
    'sacral_slope':[70.5],
    'pelvic_radius':[23.5],
    'degree_spondylolisthesis':[40.5]})
predictions = clf.predict(new_data)
predicted_labels = lb.inverse_transform(predictions)
print(f'PREDICTED LABELS: {predicted_labels}')
```
