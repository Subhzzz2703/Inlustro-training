```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
arr_txt=pd.read_csv("/content/train.csv")
```

```python
arr_txt=arr_txt.rename(columns={"Survived":"Survivers","Sex":"Gender"})
print(arr_txt)
```

```python
arr_txt.shape
```

```python
duplicate_rows_arr_txt=arr_txt[arr_txt.duplicated()]
print("Number of duplicate rows:",duplicate_rows_arr_txt.shape)
```

```python
arr_txt.count()
```

```python
print(arr_txt.isnull())
```

```python
print(arr_txt.isnull().sum())
```

```python
arr_txt=arr_txt.dropna()
arr_txt.count()
```

```python
arr_txt["Age"].fillna(arr_txt["Age"].median(),inplace=True)
arr_txt["Embarked"].fillna(arr_txt["Embarked"].mode(),inplace=True)
```

```python
#plots
import seaborn as sns
```

```python
#scatter plot
plt.figure(figsize=(8,9))
sns.scatterplot(x="PassengerId",y="Pclass",data=arr_txt,palette="muted")
plt.title("SCATTER PLOT",fontsize=15)
plt.show()
```

```python
#BAR PLOT
plt.figure(figsize=(5,9))
sns.barplot(x="Ticket",y="Fare",data=arr_txt,palette="muted")
plt.title("BAR PLOT",fontsize=15)
plt.show()
```

```python
#BOX PLOT
plt.figure(figsize=(15,9))
sns.boxplot(x="Age",y="Fare",data=arr_txt,palette="pastel")
plt.title("BOX PLOT",fontsize=15)
plt.show()
```

```python
#violin plot
plt.figure(figsize=(15,9))
sns.violinplot(x="Cabin",y="Fare",data=arr_txt,palette="muted")
plt.title("VIOLIN PLOT",fontsize=15)
plt.show()
```

```python
#line chart
plt.figure(figsize=(9,4))
sns.lineplot(x="Ticket",y="Gender",data=arr_txt,palette="dark")
plt.title("LINE PLOT",fontsize=15)
plt.show()
```

```python
#pie chart
plt.figure(figsize=(5,9))
plt.pie(x="PassengerId",data=arr_txt)
plt.title("PIE PLOT",fontsize=15)
plt.show()
```

```python
#histogram
plt.figure(figsize=(5,9))
sns.histplot(x="Ticket",y="Fare",data=arr_txt,palette="muted")
plt.title("HISTOGRAM",fontsize=15)
plt.show()
```

```python
#density plot
plt.figure(figsize=(5,9))
sns.kdeplot(x="PassengerId",y="Pclass",data=arr_txt,palette="muted")
plt.title("DENSITY PLOT",fontsize=15)
plt.show()
```

```python
#count plot
plt.figure(figsize=(5,9))
sns.countplot(x="Ticket",data=arr_txt,palette="muted")
plt.title("COUNTPLOT",fontsize=15)
plt.show()
```
